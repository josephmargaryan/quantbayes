```python
import numpy as np
```

```python
# load data
train = np.load('./train.npz')
X_train = train["X"]
y_train = train["y"].reshape((-1,1))

test = np.load('./test.npz')
X_test = test["X"]
y_test = test["y"].reshape((-1,1))

classes = {
    1: "cultivated_land",
    2: "forest",
    3: "grassland",
    4: "shrubland",
    5: "water",
    8: "artificial_surface",
    9: "bareland"
}
```

```python
print("Shape of training data: {}".format(X_train.shape))
print("Shape of training labels: {}".format(y_train.shape))
print("Shape of test data: {}".format(X_test.shape))
print("Shape of test labels: {}".format(y_test.shape))
```

The training and test set contain a few thousand instances. Each instance is based on image of size 13x13 pixels, which are available for 12 timestamps and 6 bands. That is, one is given an array of shape (12,13,13,6) for each instance. The label arrays contain the labels associated with the instances, where the central pixel/position determines the class of the (whole) image array for each instance, see below.

```python
# visualize some of the (image) data

import matplotlib.cm as cm
import matplotlib.pyplot as plt

idx = 0

print("Label (i.e., class of pixel in the center of the images) for image sequence: {}".format(classes[y_train[idx,0]]))

for year in range(12):
    fig, axs = plt.subplots(1, 6, figsize=(24, 4))
    for b in range(6):
        axs[b].imshow(X_train[idx, year, :, :, b], cmap=plt.get_cmap("Greys"))
        axs[b].set_title("Band {}".format(b+1))

```

```python
# === Complete Landcover Classification Workflow ===

# 0) Jupyter magic for inline plotting
%matplotlib inline

# 1) Suppress the scikit-learn FutureWarning about 'algorithm'
import warnings
warnings.filterwarnings(
    "ignore",
    category=FutureWarning,
    module="sklearn.ensemble._weight_boosting"
)

# 2) Imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay

# 3) Load data
train = np.load('./train.npz')
X_train = train["X"]           # (n_samples, 12, 13, 13, 6)
y_train = train["y"].ravel()   # (n_samples,)

test = np.load('./test.npz')
X_test = test["X"]
y_test = test["y"].ravel()

# 4) Feature extraction: central pixel only
def extract_central_pixel_features(X):
    n_samples, n_times, h, w, n_bands = X.shape
    ci, cj = h // 2, w // 2
    return X[:, :, ci, cj, :].reshape(n_samples, n_times * n_bands)

Xc_train = extract_central_pixel_features(X_train)  # (n_samples, 72)
Xc_test  = extract_central_pixel_features(X_test)

print("Central-pixel feature shape (train):", Xc_train.shape)
print("Central-pixel feature shape (test) :", Xc_test.shape)

# 5) Set up AdaBoost with discrete SAMME and grid search over tree depth & n_estimators
base_learners = [
    DecisionTreeClassifier(max_depth=d, random_state=0)
    for d in [1, 2, 3]
]

adb = AdaBoostClassifier(
    algorithm="SAMME",    # discrete boosting per assignment spec
    random_state=0
)

param_grid = {
    "estimator": base_learners,
    "n_estimators": [50, 100, 200]
}

grid = GridSearchCV(
    estimator=adb,
    param_grid=param_grid,
    cv=2,
    scoring="accuracy",
    n_jobs=-1,
    verbose=2,
    refit=True
)

# 6) Run grid search
grid.fit(Xc_train, y_train)

print("\n=== Grid Search Results ===")
print("Best parameters:      ", grid.best_params_)
print(f"Best CV accuracy:     {grid.best_score_:.4f}")

# 7) Evaluate on test set
best_model = grid.best_estimator_
y_pred = best_model.predict(Xc_test)
test_acc = accuracy_score(y_test, y_pred)
print(f"Test set accuracy:    {test_acc:.4f}")

# 8) Confusion matrix: display & save
labels = np.unique(y_test)
cm = confusion_matrix(y_test, y_pred, labels=labels)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=[str(l) for l in labels])

fig, ax = plt.subplots(figsize=(8, 6))
disp.plot(ax=ax, cmap=plt.cm.Blues, colorbar=True)
plt.title("Confusion Matrix on Test Set")

plt.savefig("confusion_matrix.png", dpi=300, bbox_inches="tight")
plt.show()
```
